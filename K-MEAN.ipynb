```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = {
    'cgpa' : [5.2,5.8,6.1,6.5,7.0,7.4,7.8,8.2,8.8,9.3],
    'Marks':[52,58,61,65,70,74,78,82,88,93]
}

df = pd.DataFrame(data)
df

#step 3 select feature

x = df[["cgpa","Marks"]]

#step 4 apply KMeans

kmean = KMeans(n_clusters=3,random_state=42)
df['cluster'] = kmean.fit_predict(x)

#print clustered data

print("\nclustered data:")
print(df)

print("\ncentroids")
print(kmean.cluster_centers_)

plt.scatter(df["cgpa"],df["Marks"],c=df["cluster"])
plt.scatter(kmean.cluster_centers_[:,0],kmean.cluster_centers_[:,1],marker="x",s=200)
plt.xlabel("cgpa")
plt.ylabel("Marks")
plt.title("K_MEAN CLUSTERING (STUDENTS)")
plt.show()
```
